```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/ProdigyInternship/API_SP.POP.TOTL_DS2_en_csv_v2_6508519.csv")
df.head()
```

```python
x = df.drop(['Country Code', 'Indicator Name', 'Indicator Code'], axis=1)
y = df['Country Name']
# print(x)
# print(y)
```

```python
chosen_country = 'France'
country_data = x[y == chosen_country]

years = country_data.columns[2:].astype(int)
population = country_data.iloc[:, 2:].values.flatten()
```

```python
plt.bar(years, population, color = 'pink')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {chosen_country}')

plt.show()
```

```python
plt.hist(population, bins=15, color='pink', edgecolor='black')
plt.title(f'Population of {chosen_country}')

plt.show()
```
